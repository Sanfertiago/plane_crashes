Loading and cleaning the obtained data

```python
pwd = os.getcwd() #This creates a string of the folder this Python Script is stored in

filepath = pwd + "/Data.csv" #String that is the filepath to the csv file

df = pd.read_csv(filepath, index_col=0) #We load the data file to the current session

df.columns = df.iloc[0] #we set the column labels to equal the values in the 1st row (index location 0

#df.isnull().sum() #We check if there are any null values
Data = df[df['Date'].str.contains("Date")==False] #We drop the repeated "index" value rows by filtering the strings containg the "Date" value


#We split the column fatalities into Deaths and Passengers respectivedly
Data[['Deaths', 'Passengers0']] = Data['Fatalities'].str.split('/', expand=True)
Data[['Passengers','Drop']] = Data['Passengers0'].str.split('(', expand=True)
Data.drop(columns=['Fatalities', 'Passengers0', 'Drop'])


```
